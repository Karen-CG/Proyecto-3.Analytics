### 2.Tranformaciones y normalización del dataset

Aplicando las transformaciones para la columna de 'fecha'

```python
#Cambiando el formato de la columna
data['fecha']=pd.to_datetime(data['fecha'])
```

Normalización y tranformación de la columna 'HORA declarada'

```python
# Normalizando la columna de tiempo (HORA declarada)
data['HORA declarada'] = data ['HORA declarada'].replace('?', '00:00')
data['HORA declarada'] = data ['HORA declarada'].str.replace('c ','')
```

```python
#Visualización de los cambios de la columna HORA
data
```

```python
# Verificando cambios en las columnas  que presentaban el formato 'c '
# data.iloc[95]
# data.iloc[117]
data.iloc[169]
```

```python
# Creando función para normalizar la columna de HORA declarada
def normalizar_hora(var):
    if ':' not in var:
        return f'{var[:2]}:{var[-2:]}'
    else:
        return var
```

```python
# Verificando la función normalizar_hora
var1= '1550'
var2 =normalizar_hora(var1)
var2
```

```python
var1= '16:80'
var2 =normalizar_hora(var1)
var2
```

```python
# data['HORA declarada'] = data ['HORA declarada'].astype(str)
data['HORA declarada'] = data ['HORA declarada'].apply(lambda x: normalizar_hora(x))
```

```python
#Verificando cambios en la columna 'HORA declarada'
data
```

```python
# Reemplazando valores  donde tenemos el signo ? por Na
data.replace({'?': 'Na'}, inplace= True)
```

```python
# Verificando cambios en el dataset
data
```

```python
# Para las columnas donde tenemos valores numericos reemplazaremos por cero en lugar de Na
data['all_aboard'] = data ['all_aboard'].replace('Na', 0)
data['PASAJEROS A BORDO'] = data ['PASAJEROS A BORDO'].replace('Na', 0)
data['crew_aboard'] = data ['crew_aboard'].replace('Na', 0)
data['cantidad de fallecidos'] = data ['cantidad de fallecidos'].replace('Na', 0)
data['passenger_fatalities'] = data ['passenger_fatalities'].replace('Na', 0)
data['crew_fatalities'] = data ['crew_fatalities'].replace('Na', 0)
data['ground'] = data ['ground'].replace('Na', 0)
```
