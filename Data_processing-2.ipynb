### 2. Normalizando y transformando el dataset

Se realizara una reducción de la dimensionalidad del dataset y se tomaran las columnas que pueden aportar información  necesaria en el análisis de las causas de la caída en los aviones.

```python
# Reduciendo la cantidad de columnas tomadas del data set
aviation_data = aviation_data[['Event Id ', ' Investigation Type ', ' Event Date ', ' Location ', ' Country ', ' Latitude ', ' Longitude ',' Airport Name ', ' Injury Severity ', ' Aircraft Category ', ' Make ',' Model ',' Purpose of Flight ',
        ' Air Carrier ', ' Total Fatal Injuries ', ' Total Serious Injuries ', ' Total Minor Injuries ',
       ' Total Uninjured ', ' Weather Condition ', ' Broad Phase of Flight ']]
```

```python
# Cambiando formato de la columna de ´' Event Date ' a datatime
aviation_data[' Event Date ']= pd.to_datetime(aviation_data[' Event Date '])
```

```python
#Normaizando los nombres de las columnas
aviation_data =aviation_data.rename(columns={'Event Id ':'Event Id' })
aviation_data =aviation_data.rename(columns={' Investigation Type ':'Investigation Type' })
aviation_data =aviation_data.rename(columns={' Event Date ':'Event Date' })
aviation_data =aviation_data.rename(columns={' Location ':'Location' })
aviation_data =aviation_data.rename(columns={' Country ':'Country' })
aviation_data =aviation_data.rename(columns={' Latitude ':'Latitude' })
aviation_data =aviation_data.rename(columns={' Longitude ':'Longitude' })
aviation_data =aviation_data.rename(columns={' Airport Name ':'Airport Name' })
aviation_data =aviation_data.rename(columns={' Injury Severity ':'Injury Severity' })
aviation_data =aviation_data.rename(columns={' Aircraft Category ':'Aircraft Category' })
aviation_data =aviation_data.rename(columns={' Make ':'Make' })
aviation_data =aviation_data.rename(columns={' Model ':'Model' })
aviation_data =aviation_data.rename(columns={' Purpose of Flight ':'Purpose of Flight' })
aviation_data =aviation_data.rename(columns={' Air Carrier ':'Air Carrier' })
aviation_data =aviation_data.rename(columns={' Total Fatal Injuries ':'Total Fatal Injuries'})
aviation_data =aviation_data.rename(columns={' Total Serious Injuries ':'Total Serious Injuries'})
aviation_data =aviation_data.rename(columns={' Total Minor Injuries ':'Total Minor Injuries'})
aviation_data =aviation_data.rename(columns={' Total Uninjured ':'Total Uninjured'})
aviation_data =aviation_data.rename(columns={' Weather Condition ':'Weather Condition'})
aviation_data =aviation_data.rename(columns={' Broad Phase of Flight ':'Broad Phase of Flight'})
```

```python
# Verificando los cambios en la estructura del dataset
aviation_data
```

```python
# Reemplezando los datos que se encuentran vacios por etiquetas que puedan brindar mayor información
aviation_data['Investigation Type'] = aviation_data['Investigation Type'].replace('  ','No information')
aviation_data['Location'] = aviation_data['Location'].replace('  ','Unknown')
aviation_data['Country'] = aviation_data['Country'].replace('  ','Unknown')
# aviation_data['Latitude'] = aviation_data['Latitude'].replace('  ','Na')
# aviation_data['Longitude'] = aviation_data['Longitude'].replace('  ','Na')
aviation_data['Airport Name'] = aviation_data['Airport Name'].replace('  ','No information')
aviation_data['Aircraft Category'] = aviation_data['Aircraft Category'].replace('  ','Unknown')
aviation_data['Make'] = aviation_data['Make'].replace('  ','Unknown')
```

```python
#En la columna de 'Make' se normaliza la forma en la que vienen lso string
aviation_data['Make'] = aviation_data['Make'].str.title()
```

```python
# Cambiando formato de tipo objetc a int, tambien se reemplazan valores vacios por cero
aviation_data['Total Fatal Injuries'] = aviation_data['Total Fatal Injuries'].replace('  ',0)
aviation_data['Total Fatal Injuries'] = aviation_data['Total Fatal Injuries'].astype(int)

aviation_data['Total Serious Injuries'] = aviation_data['Total Serious Injuries'].replace('  ',0)
aviation_data['Total Serious Injuries'] = aviation_data['Total Serious Injuries'].astype(int)

aviation_data['Total Minor Injuries'] = aviation_data['Total Minor Injuries'].replace('  ',0)
aviation_data['Total Minor Injuries'] = aviation_data['Total Minor Injuries'].astype(int)

aviation_data['Total Uninjured'] = aviation_data['Total Uninjured'].replace('  ',0)
aviation_data['Total Uninjured'] = aviation_data['Total Uninjured'].astype(int)
```

```python
# Se verifica columna 'Injury Severity'
aviation_data['Injury Severity']
```

```python
# Se realiza la separación de los datos, para solo obtener la etiqueta 'Faltal'
new= aviation_data['Injury Severity'].str.split("(", n = 1, expand = True)
```

```python
# Resultado de columna 'Injury Severity'
new
```

```python
# Se asignan valores correctos a la columna 'Injury Severity'
aviation_data['Injury Severity']= new [0]
```
